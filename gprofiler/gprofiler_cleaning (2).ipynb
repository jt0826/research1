```python
import pandas as pd 
```

```python
LumA = pd.read_csv(r'C:\Users\javte\Documents\SSEF project\datasets\gprofiler\gProfiler_LumA_5000tree.csv')
LumB = pd.read_csv(r'C:\Users\javte\Documents\SSEF project\datasets\gprofiler\gProfiler_LumB.csv')
her2 = pd.read_csv(r'C:\Users\javte\Documents\SSEF project\datasets\gprofiler\gProfiler_her2.csv')
basal = pd.read_csv(r'C:\Users\javte\Documents\SSEF project\datasets\gprofiler\gProfiler_basal.csv')
normal = pd.read_csv(r'C:\Users\javte\Documents\SSEF project\datasets\gprofiler\gProfiler_normal.csv')
```

```python
def filtered_pathways(gprofiler):
    gprofiler.query('adjusted_p_value <= 0.01 and source == "GO:BP"', inplace=True)              
    return gprofiler
```

```python
types = (LumA, LumB, her2, basal, normal)
names = ('LumA', 'LumB', 'her2', 'basal', 'normal')
for type, name in zip(types, names):
    filtered_pathways(type).to_csv(rf'C:\Users\javte\Documents\SSEF project\datasets\gprofiler\cleaned_gProfiler\{name}.csv')
```
