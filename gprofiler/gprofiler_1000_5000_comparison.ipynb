```python
import pandas as pd
```

```python
data_tree1000 = pd.read_csv(r'/home/bpm_javier/Documents/results/gprofiler/gProfiler_LumA_1000tree.csv')
data_tree1000.query('source=="GO:BP",term_size<=500,adjusted_p',inplace=True)
tree1000 = data_tree1000.iloc[:10,[1,4]]
print(tree1000)
```

```python
data_tree5000 = pd.read_csv(r'/home/bpm_javier/Documents/results/gprofiler/gProfiler_LumA_5000tree.csv')
data_tree5000.query('source=="GO:BP"',inplace=True)
tree5000 = data_tree5000.iloc[:10, [1,4]]
print(tree5000)
```

```python
import matplotlib.pyplot as plt
import numpy as np
tree5000.plot.barh(x='term_name', title='-lg p for 5000tree')
plt.legend(["-lg p"]);
plt.savefig('top10_pval_5000tree.png', bbox_inches='tight')

plt.show()
```

```python
tree1000.plot.barh(x='term_name', title='-lg p for 1000tree')
plt.legend(["-lg p"]);
plt.savefig('top10_pval_1000tree.png', bbox_inches='tight')
```

```python
from matplotlib_venn import venn2
from matplotlib_venn import venn3
plt.style.use('seaborn')

plt.figure(figsize=(10,10))
plt.title("Venn Diagram For GOBP gene1000tree vs gen5000tree")

venn2([set(data_tree1000.term_name), 
       set(data_tree5000.term_name)],
       set_labels=('GOBP - 1000 trees', 'GOBP - 5000 trees')
     )

plt.show()
```
