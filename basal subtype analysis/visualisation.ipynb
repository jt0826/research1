```python
import pandas as pd
lower = 0
upper = 20
basal1 = pd.read_csv(r"C:\Users\javte\Downloads\New folder\basal1 gprofiler.csv")
basal4 = pd.read_csv(r"C:\Users\javte\Downloads\New folder\basal4 gprofiler.csv")
basal7 = pd.read_csv(r"C:\Users\javte\Downloads\New folder\basal7 gprofiler.csv")
original_basal = pd.read_csv(r"C:\Users\javte\Documents\SSEF project\datasets\gprofiler\gprofiler_2\filtered\basal.csv")
term_names = set(basal1.term_name[lower:upper]).union(set(basal4.term_name[lower:upper]),
                                                    set(basal7.term_name[lower:upper]),
                                                    set(original_basal.term_name[lower:upper]))
```

```python
df = pd.DataFrame(index=list(term_names))
basal1 = basal1[['term_name','negative_log10_of_adjusted_p_value']].iloc[lower:upper,:]
basal4 = basal4[['term_name','negative_log10_of_adjusted_p_value']].iloc[lower:upper,:]
basal7 = basal7[['term_name','negative_log10_of_adjusted_p_value']].iloc[lower:upper,:]
original_basal=original_basal[['term_name','negative_log10_of_adjusted_p_value']].iloc[lower:upper,:]
```

```python
basal1.index=basal1.term_name

basal4.index=basal4.term_name

basal7.index=basal7.term_name
original_basal.index=original_basal.term_name
del basal1['term_name']
del basal4['term_name']
del basal7['term_name']
del original_basal['term_name']
```

```python
import numpy as np
df = df.join(original_basal)
df = df.join(basal1,lsuffix='_1',rsuffix='_2')
df = df.join(basal4)
df = df.join(basal7,lsuffix='_3',rsuffix='_4')
```

```python
df=pd.DataFrame(data=np.array(df),columns=[['original_basal','basal1','basal4','basal7']],index=list(term_names))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig,axes = plt.subplots(1,1,dpi=800,figsize=(16,16))
sns.heatmap(df, square=True, annot = True,annot_kws={'size': 10}, cbar_kws={'shrink': 0.5, 'label': "-lg p"},cmap='YlGn')
```
