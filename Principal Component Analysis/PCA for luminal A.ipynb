```python
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
#transpose matrix to have sample as row, gene in column
LumA_transpose = pd.read_csv(r'C:\Users\javte\Documents\SSEF project\datasets\gene expression+metadata\LumA_NT_logCPMrmBatch_20221128.csv', index_col=[0])
LumA = LumA_transpose.transpose()
LumA.head()
```

```python
#PCA
pca = PCA(n_components=50)
principalComponents_LumA = pca.fit_transform(LumA)
principal_LumA_Df = pd.DataFrame(data = principalComponents_LumA
             , columns = ["PC"+str(i) for i in range(1, 51)])

 
```

```python
principal_LumA_Df.head()
```

```python
# Take pam50 column from metadata and set as index of PCA dataframe
LumA_metadata = pd.read_csv(r"C:\Users\javte\Documents\SSEF project\datasets\gene expression+metadata\LumA_NT_metadat_20221128.csv")
LumA_pam50 = LumA_metadata[('pam50')].to_list()
principal_LumA_Df['pam50'] = LumA_pam50
principal_LumA_Df.set_index('pam50', drop = True, inplace = True)
principal_LumA_Df.head()
```

```python
#formatting plot
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('PC1',fontsize=20)
plt.ylabel('PC2',fontsize=20)
plt.title("LumA PCA",fontsize=20)
targets = ['Healthy', 'LumA']
colours = ['r', 'g']
#colour code points according to pam50
for target, colour in zip(targets,colours):
    index = principal_LumA_Df.index == target
    plt.scatter(principal_LumA_Df.loc[index, 'PC1']
               , principal_LumA_Df.loc[index, 'PC2'], c = colour, s = 50)
#plot legend
plt.legend(targets,prop={'size': 15})
```

```python
import matplotlib.pyplot as plt
#no. of PCs to be visualised
n = 10
plt.bar(x=principal_LumA_Df.columns[:n], height = 100*pca.explained_variance_ratio_[:n])
cumulative = np.cumsum(100*pca.explained_variance_ratio_[:n])
plt.plot(principal_LumA_Df.columns[:n], cumulative)
cumulative = np.cumsum(pca.explained_variance_ratio_[:n])
plt.xlabel('principal component')
plt.ylabel('explained variance (%)')
plt.legend(['cumulative'],prop={'size': 10},loc='best')
```
