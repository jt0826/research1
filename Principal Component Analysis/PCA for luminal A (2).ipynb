```python
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#transpose matrix
LumA_transpose = pd.read_csv(r'C:\Users\javte\Documents\SSEF project\Normalized expression data set\LumA_NT_logCPMrmBatch_20221128.csv', index_col=[0])
LumA = LumA_transpose.transpose()
LumA.reset_index(inplace=True)
del LumA['index']
X = LumA
```

```python
LumA_metadata = pd.read_csv(r"C:\Users\javte\Documents\SSEF project\Normalized expression data set\metadata\LumA_NT_metadat_20221128.csv")
LumA_metadata.head()
LumA_metadata[['pam50']]
# convert 'healthy' = 0; 'lumA' = 1
LumA_metadata['pam50'].replace('Healthy', 0, inplace=True)
LumA_metadata['pam50'].replace('LumA', 1, inplace=True)
LumA_metadata.head
y = LumA_metadata['pam50']
print(y)
from sklearn.model_selection import train_test_split
# Split into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = RandomForestClassifier(n_estimators=100, 
                               max_depth=5, 
                               random_state=77, 
                               verbose=2, 
                               bootstrap=True, 
                               oob_score=True, 
                               n_jobs=-1)
```

```python
feat_selector = BorutaPy(
    verbose=5,
    estimator=model,
    n_estimators=5000,
    alpha = 0.0001,
    max_iter=1000  
)
```
